# 1. Distribuição de frequência

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from collections import Counter

%matplotlib inline

sns.set()
```

```python
enem_sp = pd.read_csv('/content/enem_2019_tratado.csv',sep=',',encoding='iso-8859-1')
```

```python
enem_sp.head()
```

```python
enem_sorocaba = enem_sp.loc[enem_sp['NO_MUNICIPIO_RESIDENCIA']=='Sorocaba']
```

```python
enem_sorocaba.head()
```

```python
idades = enem_sorocaba['IDADE']
```

```python
idades
```

```python
freq_absoluta = Counter(idades)
freq_absoluta
```

```python
freq_absoluta = pd.DataFrame.from_dict(freq_absoluta, orient='index')
```

```python
freq_absoluta = freq_absoluta.sort_index()
freq_absoluta
```

```python
freq_relativa = freq_absoluta/freq_absoluta.sum()
freq_relativa
```

```python
freq_relativa_perc = freq_relativa*100
freq_relativa_perc
```

# 2. Histograma

#### Matplotlib

```python
plt.figure(figsize=(20,12))
plt.title('Distribuição de Idades - ENEM 2019 - Sorocaba/SP')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(x='IDADE', data=enem_sorocaba, bins=10, color='red', density=True);
```

#### Seaborn

```python
plt.figure(figsize=(20,12))
sns.distplot(enem_sorocaba['IDADE'], bins=8, kde=True);
```

#### Plotly

```python
grafico = px.histogram(enem_sorocaba, x='IDADE', nbins=10, histnorm='percent')
grafico.update_layout(width=1000, height=700, title_text='Distibuição de idades - ENEM 2019 - Sorocaba')
grafico.show()
```

```python
grafico = px.histogram(enem_sorocaba, x='IDADE', nbins=10, histnorm='percent',color='SEXO')
grafico.update_layout(width=1000, height=700, title_text='Distibuição de idades - ENEM 2019 - Sorocaba')
grafico.show()
```

```python
grafico = px.histogram(enem_sorocaba, x='NOTA_MT', nbins=20, histnorm='percent',color='SEXO')
grafico.update_layout(width=1000, height=700, title_text='Notas de Matemática por gênero - ENEM 2019 - Sorocaba')
grafico.show()
```

```python
grafico = px.histogram(enem_sorocaba, x='NOTA_CN', nbins=20, histnorm='percent',color='SEXO')
grafico.update_layout(width=1000, height=700, title_text='Notas de Matemática por gênero - ENEM 2019 - Sorocaba')
grafico.show()
```
