## Deletando colunas

```python
df = df.drop('lat', axis=1)

df = df.drop('long', axis=1)

# df = df.drop(['lat', 'long'], axis=1) # no sentido da coluna
```

## Criando coluna

```python
df['teste'] = 'teste'
```

```python
df['teste'] = df['sqft_lot'] - df['sqft_living']

df['teste'] = df['sqft_lot'] + df['sqft_living']

df['teste'] = df['sqft_lot'] / df['sqft_living']

df = df.drop('teste', axis=1)
```

### Criando coluna de classe

```python
#1.Qual a quantidade de imóveis por nível?

# Nível 0: Preço entre RS0 e RS300.000;
# Nível 1: Preço entre RS300.000 e RS450.00;
# Nível 2: Preço entre RS450.000 e RS650.000;
# Nível 3: Preço acima de RS650.000.



for i in range(len(df)):
    if (df.loc[i,'price'] >= 0) & (df.loc[i, 'price'] < 300000):
        df.loc[i, 'Nivel'] = 'NIVEL 0'
    
    elif (df.loc[i,'price'] >= 300000) & (df.loc[i, 'price'] < 450000):
        df.loc[i, 'Nivel'] = 'NIVEL 1'
        
    elif (df.loc[i,'price'] >= 450000) & (df.loc[i, 'price'] < 650000):
        df.loc[i, 'Nivel'] = 'NIVEL 2'
        
    else:
        df.loc[i, 'Nivel'] = 'NIVEL 3'
```

## Countplot

```python
plt.figure(figsize=(20,10))
grafico = sns.countplot(x='Nivel', data=df, order = df['Nivel'].value_counts().index, palette='inferno')

ax=grafico

for p in ax.patches:
             ax.annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='black', xytext=(0, 10),
                 textcoords='offset points')
        
plt.title('Frequência de imoveis', fontsize=18)
plt.xlabel('Níveis', fontsize=16)
plt.ylabel('Quantidade de imoveis', fontsize=16)

plt.savefig('grafico.png')
```
